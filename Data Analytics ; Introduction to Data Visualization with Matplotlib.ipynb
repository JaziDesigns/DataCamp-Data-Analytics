### Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Using the matplotlib.pyplot interface

```python
# Create a Figure and an Axes with plt.subplots
fig, ax = plt.subplots(figsize = (18, 7))

# Call the show function to show the result
plt.xticks(size = 15)
plt.yticks(size = 15)
plt.show()
```

**Seattle Data**

```python
seattle = pd.read_csv("seattle_weather.csv")
seattle[:2]
```

**Average rainfall in each month during a ten-year period. (Seattle)**

```python
srm = seattle.groupby("DATE")[["MLY-PRCP-NORMAL", "MLY-TAVG-NORMAL", "MLY-TAVG-STDDEV", "MLY-PRCP-25PCTL", "MLY-PRCP-75PCTL"]].agg(["mean"]).round(2).reset_index()
srm.columns = ["Month", "Seattle Rainfall", "Seattle Temperature", "Seattle Temperature Standard Deviation", "Seattle Rainfall 25%", "Seattle Rainfall 75%"]
srm
```

**Assign Month Names by using Mapping**

```python
Month = {1: "January",
         2: "February",
         3: "March",
         4: "April",
         5: "May",
         6: "June",
         7: "July",
         8: "August",
         9: "September",
         10: "October",
         11: "November",
         12: "December"}
```

```python
srm["Month"] = srm["Month"].map(Month)
srm
```

**Austin Data**

```python
austin = pd.read_csv("austin_weather.csv")
austin[:2]
```

**Average rainfall in each month during a ten-year period. (Austin)**

```python
arm = austin.groupby("DATE")[["MLY-PRCP-NORMAL", "MLY-TAVG-NORMAL", "MLY-TAVG-STDDEV", "MLY-PRCP-25PCTL", "MLY-PRCP-75PCTL"]].agg(["mean"]).reset_index()
arm.columns = ["Month", "Austin Rainfall", "Austin Temperature", "Austin Temperature Standard Deviation", "Austin Rainfall 25%", "Austin Rainfall 75%"]
arm
```

```python
arm["Month"] = arm["Month"].map(Month)
arm
```

### Adding data to an Axes object

```python
fig, ax = plt.subplots(figsize = (18, 7))

ax.plot(srm["Month"], srm["Seattle Rainfall"], label = "Seattle Rainfall", linewidth = 3, color = "blue", marker = "o", linestyle = "--")
ax.plot(arm["Month"], arm["Austin Rainfall"], label = "Austin Rainfall", linewidth = 3, color = "red", marker = "o", linestyle = "--")
plt.title("Average rainfall by month in a ten years period", size = 15)
plt.legend(loc = 9, fontsize = 15)
plt.ylabel("Average Rainfall", size = 15)
plt.xticks(size = 15)
plt.yticks(size = 15)
plt.show()
```

### Creating small multiples with plt.subplots

```python
fig, ax = plt.subplots(2, 2, figsize = (18, 7))
fig.suptitle("Rainfall & Temperature Comparison of Seattle & Austin", fontsize = 15)

# Addressing the top left Axes as index 0, 0, plot month and Seattle precipitation
ax[0, 0].plot(srm["Month"], srm["Seattle Rainfall"], color = "blue")
plt.sca(ax[0, 0])
plt.xticks(srm["Month"], ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"])
plt.ylabel("Seattle Rainfall")

# In the top right (index 0,1), plot month and Seattle temperatures
ax[0, 1].plot(srm["Month"], srm["Seattle Temperature"], color = "red")
plt.sca(ax[0, 1])
plt.xticks(srm["Month"], ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"])
plt.ylabel("Seattle Temperature")

# In the bottom left (1, 0) plot month and Austin precipitations
ax[1, 0].plot(arm["Month"], arm["Austin Rainfall"], color = "blue")
plt.sca(ax[1, 0])
plt.xticks(arm["Month"], ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"])
plt.ylabel("Austin Rainfall")

# In the bottom right (1, 1) plot month and Austin temperatures
ax[1, 1].plot(arm["Month"], arm["Austin Temperature"], color = "red")
plt.sca(ax[1, 1])
plt.xticks(arm["Month"], ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"])
plt.ylabel("Austin Temperature")

print("\n")
plt.show()
```

### Small multiples with shared y axis

```python
fig, ax = plt.subplots(2, 1, sharey = True, figsize = (18, 7))

fig.suptitle("Rainfall Comparison of Seattle & Austin", fontsize = 15)

ax[0].plot(srm["Month"], srm["Seattle Rainfall 25%"], color = "blue", linestyle = "--")
ax[0].plot(srm["Month"], srm["Seattle Rainfall"], color = "blue")
ax[0].plot(srm["Month"], srm["Seattle Rainfall 75%"], color = "blue", linestyle = "--")
plt.sca(ax[0])
plt.ylabel("Seattle Rainfall")

ax[1].plot(arm["Month"], arm["Austin Rainfall 25%"], color = "red", linestyle = "--")
ax[1].plot(arm["Month"], arm["Austin Rainfall"], color = "red")
ax[1].plot(arm["Month"], arm["Austin Rainfall 75%"], color = "red", linestyle = "--")
plt.sca(ax[1])
plt.ylabel("Austin Rainfall")

plt.show()
```

### Read data with a time index

```python
df = pd.read_csv("climate_change.csv", parse_dates = ["date"], index_col = "date")
df[:5]
```

```python
fig, ax = plt.subplots(figsize = (18, 7))
ax.plot(df.index, df["relative_temp"], color = "red")
ax.set_title("Relative Temperature (Celcius) by Date", size = 15)
plt.xticks(size = 15)
plt.yticks(size = 15)
plt.show()
```

### Using a time index to zoom in

```python
fig, ax = plt.subplots(figsize = (18, 7))

eightees = df["1980-01-01": "1988-12-31"]

ax.plot(eightees.index, eightees["co2"], color = "orange", linewidth = 3)
ax.set_title("Co2 in Eightees", size = 15)
ax.set_ylabel("Co2", size = 15)
plt.xticks(size = 15)
plt.yticks(size = 15)
plt.show()
```

### Plotting two variables

```python
fig, ax = plt.subplots(figsize = (18, 7))

ax.plot(df.index, df["co2"], color = "blue")
ax.set_title("Co2 and Temperature trends by Years", size = 15)
ax.set_xlabel("Time",size = 15)
ax.set_ylabel("Co2 (ppm)", size = 15, color = "blue")
ax.tick_params("y", colors = "blue", labelsize = 15)
ax.tick_params("x", labelsize = 15)


ax2 = ax.twinx()

ax2.plot(df.index, df["relative_temp"], color = "red")
ax2.set_ylabel("Relative Temperature (Celcius)", size = 15, color = "red")
ax2.tick_params("y", colors = "red", labelsize = 15)
plt.show()
```

### Annotating a plot of time-series data

```python
fig, ax = plt.subplots(figsize = (18, 7))

ax.plot(df.index, df["relative_temp"])

ax.annotate(">1 degree", fontsize = 20, xy = (pd.Timestamp('2015-10-06'), 1))

ax.tick_params('y', labelsize = 15)
ax.tick_params('x', labelsize = 15)
plt.show()
```

### Plotting time-series: putting it all together

```python
fig, ax = plt.subplots(figsize = (18, 7))

ax.plot(df.index, df["co2"], color = "blue")
ax.tick_params('y', labelsize = 15, colors = "blue")
ax.tick_params('x', labelsize = 15)
ax.set_ylabel("CO2 levels", size = 15, color = "blue")
ax.set_title("Time Series Graph by CO2 levels and Relative temp (Celsius)", size = 15)

ax2 = ax.twinx()

ax2.plot(df.index, df["relative_temp"], color = "red")
ax2.tick_params('y', labelsize = 15, colors = "red")
ax2.set_ylabel("Relative temp (Celsius)", color = "red", size = 15)
ax2.annotate(">1 Degree", fontsize = 20, xy = (pd.Timestamp('2015-10-06'), 1),
             xytext = (pd.Timestamp('2008-10-06'), -0.2), arrowprops = {'arrowstyle': '->', "color": "black"})

fig.set_size_inches([16, 9])
fig.savefig("Co2_RelativeTemp.png")
plt.show()
```

### Bar Chart

```python
medals = pd.read_csv("medals_by_country_2016.csv")
medals = medals.rename(columns = {"Unnamed: 0": "Country"})
medals[:5]
```

```python
fig, ax = plt.subplots(figsize = (18, 7))
ax.bar(medals["Country"], medals["Gold"], color = "deeppink")
ax.set_title("Gold Medals by Countries in 2016 Olympics", color = "red", size = 20)
ax.set_ylabel("Gold Medals", color = "red", size = 20)
ax.tick_params('x', labelsize= 14)
ax.tick_params('y', labelsize= 14)
plt.show()
```

### Stacked bar chart

```python
fig, ax = plt.subplots(figsize = (18, 7))

ax.bar(medals["Country"], medals["Gold"], label = "Gold")
ax.bar(medals["Country"], medals["Silver"], bottom = medals["Gold"], label = "Silver")
ax.bar(medals["Country"], medals["Bronze"], bottom = medals["Gold"] + medals["Silver"], label = "Bronze")
ax.set_title("Gold, Silver & Bronze Medals by Country", size = 20)
ax.set_ylabel("Medals", size = 20)
ax.tick_params('x', labelsize = 14)
ax.tick_params('y', labelsize = 14)


ax.legend()
plt.show()
```

### Histograms

```python
olympics = pd.read_csv("summer2016.csv")
olympics[:5]
```

**Extract Men's Rowing Games**

```python
men_rowing = olympics[(olympics["Sport"] == "Rowing") & (olympics["Sex"] == "M")]
men_rowing[:2]
```

**Extract Men's Gymnasium Games**

```python
men_gymnastics = olympics[(olympics["Sport"] == "Gymnastics") & (olympics["Sex"] == "M")]
men_gymnastics[:2]                                                            
```

```python
fig, ax = plt.subplots(figsize = (18, 7))

ax.hist(men_rowing["Weight"], label = "Rowing", bins = 30, alpha = 0.7)
ax.hist(men_gymnastics["Weight"], label = "Gymnastics", bins = 10, alpha = 0.7)
ax.set_title("Distribution of Rowing & Gymnastics Players by Weight", size = 20, color = "red")
ax.set_xlabel("Weight", size = 20, color = "red")
ax.set_ylabel("Distribution", size = 20, color = "red")
ax.tick_params('x', labelsize = 15)
ax.tick_params('y', labelsize = 15)
ax.legend()
plt.show()
```

### Step Histogram

```python
fig, ax = plt.subplots(figsize = (18, 7))

ax.hist(men_rowing["Weight"], label = "Rowing", bins = 30, histtype = "step", linewidth = 3)
ax.hist(men_gymnastics["Weight"], label = "Gymnastics", bins = 10, histtype = "step", linewidth = 3)
ax.set_title("Distribution of Rowing & Gymnastics Players by Weight (Step Histogram)", size = 20, color = "red")
ax.set_xlabel("Weight", size = 20, color = "red")
ax.set_ylabel("Distribution", size = 20, color = "red")
ax.tick_params('x', labelsize = 15)
ax.tick_params('y', labelsize = 15)
ax.legend()
plt.show()
```

### Statistical Plots

**Error Bars on Bar Chart**

```python
fig, ax = plt.subplots(figsize = (18, 7))

ax.bar("Rowing", men_rowing["Height"].mean(), yerr = men_rowing["Height"].std())
ax.bar("Gymnastics", men_gymnastics["Height"].mean(), yerr = men_rowing["Height"].std())
ax.set_title("Error Bars of Rowing & Gymnastics Players by Height)", size = 20, color = "red")
ax.set_xlabel("Weight", size = 20, color = "red")
ax.set_ylabel("Average", size = 20, color = "red")
ax.tick_params('x', labelsize = 15)
ax.tick_params('y', labelsize = 15)
plt.show()
```

**Error Bars on Line Plot**

```python
fig, ax = plt.subplots(figsize = (18, 7))

ax.errorbar(srm["Month"], srm["Seattle Temperature"], yerr = srm["Seattle Temperature Standard Deviation"], color = "blue", label = "Seattle", linewidth = 3)
ax.errorbar(arm["Month"], arm["Austin Temperature"], yerr = arm["Austin Temperature Standard Deviation"], color = "red", label = "Austin", linewidth = 3)
ax.set_title("Error Bars of Seattle & Austin Temperature)", size = 20, color = "green")
ax.set_ylabel("Temperature (Fahrenheit)", size = 20, color = "green")
ax.tick_params('x', labelsize = 15)
ax.tick_params('y', labelsize = 15)
ax.legend(fontsize = 15)
plt.show()
```

### Boxplots

```python
fig, ax = plt.subplots(figsize = (18, 7))

ax.boxplot([men_rowing["Height"], men_gymnastics["Height"]])
ax.set_xticklabels(["Rowing", "Gymnastics"], size = 15)
ax.tick_params('y', labelsize = 15)
ax.set_ylabel("Height (cm)", size = 15)
ax.set_title("Boxplots of Rowing and Gymnastic Athlete's Height", size = 20, color = "deeppink")
plt.show()
```

### Weight of Male Athletes by Sports

```python
male_weights = male.groupby("Sport")["Weight"].agg(["mean", "std"]).reset_index()
male_weights.columns = ["Sport", "Average Male Weight", "Standard Deviation"]

fig, ax = plt.subplots(figsize = (18, 7))

ax.bar(male_weights["Sport"], male_weights["Average Male Weight"], yerr = male_weights["Standard Deviation"], color = "orange")
ax.set_title("Average Weight of Male Athletes by Sports", size = 20, color = "brown")
ax.tick_params('x', rotation = 90, labelsize = 15)
ax.tick_params('y', labelsize = 15)
ax.set_ylabel("Average Weight (Men)", size = 20, color = "brown")
plt.show()
```
