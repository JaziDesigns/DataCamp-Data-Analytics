### Import Libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from numpy import median
```

### Scatter Plots

```python
gdp = pd.read_csv("countries-of-the-world.csv", skipinitialspace = True)
gdp[:3]
```

```python
gdp.info()
```

```python
gdp = gdp.dropna()
```

```python
gdp["Pop. Density (per sq. mi.)"] = gdp["Pop. Density (per sq. mi.)"].apply(lambda x: x.replace(",", ".")).astype("float")

gdp["Coastline (coast/area ratio)"] = gdp["Coastline (coast/area ratio)"].apply(lambda x: x.replace(",", ".")).astype("float")

gdp["Net migration"] = gdp["Net migration"].apply(lambda x: x.replace(",", ".")).astype("float")

gdp["Infant mortality (per 1000 births)"] = gdp["Infant mortality (per 1000 births)"].apply(lambda x: x.replace(",", ".")).astype("float")

gdp["Literacy (%)"] = gdp["Literacy (%)"].apply(lambda x: x.replace(",", ".")).astype("float")

gdp["Phones (per 1000)"] = gdp["Phones (per 1000)"].apply(lambda x: x.replace(",", ".")).astype("float")

gdp["Arable (%)"] = gdp["Arable (%)"].apply(lambda x: x.replace(",", ".")).astype("float")

gdp["Crops (%)"] = gdp["Crops (%)"].apply(lambda x: x.replace(",", ".")).astype("float")

gdp["Other (%)"] = gdp["Other (%)"].apply(lambda x: x.replace(",", ".")).astype("float")

gdp["Climate"] = gdp["Climate"].apply(lambda x: x.replace(",", ".")).astype("float")

gdp["Birthrate"] = gdp["Birthrate"].apply(lambda x: x.replace(",", ".")).astype("float")

gdp["Deathrate"] = gdp["Deathrate"].apply(lambda x: x.replace(",", ".")).astype("float")

gdp["Agriculture"] = gdp["Agriculture"].apply(lambda x: x.replace(",", ".")).astype("float")

gdp["Industry"] = gdp["Industry"].apply(lambda x: x.replace(",", ".")).astype("float")

gdp["Service"] = gdp["Service"].apply(lambda x: x.replace(",", ".")).astype("float")
```

```python
gdp[:2]
```

```python
gdp.info()
```

**Phones (per 1000) by GDP ($ per capita)**

```python
# Create scatter plot with GDP on the x-axis and number of phones on the y-axis

plt.figure(figsize = (18, 7))
sns.scatterplot(x = "GDP ($ per capita)", y = "Phones (per 1000)", data = gdp, s = 100, alpha = 0.5, color = "red")
plt.title("Phones (per 1000) by GDP ($ per capita)", size = 20, color = "blue")
plt.xticks(size = 15)
plt.yticks(size = 15)
plt.xlabel("GDP ($ per capita)", size = 20, color = "blue")
plt.ylabel("Phones (per 1000)", size = 20, color = "blue")
plt.show()
```

**Literacy (%) by GDP ($ per capita)**

```python
# Create scatter plot with GDP on the x-axis and Literacy on the y-axis

plt.figure(figsize = (18, 7))
sns.scatterplot(x = "GDP ($ per capita)", y = "Literacy (%)", data = gdp, s = 100, color = "orange")
plt.title("Literacy (%) by GDP ($ per capita)", size = 20, color = "brown")
plt.xticks(size = 15)
plt.yticks(size = 15)
plt.xlabel("GDP ($ per capita)", size = 20, color = "brown")
plt.ylabel("Literacy (%)", size = 20, color = "brown")
plt.show()
```

**Number of Countries by Region**

```python
gdp["Region"] = gdp["Region"].str.strip()
```

```python
plt.figure(figsize = (18, 7))
sns.countplot(x = "Region", data = gdp)
plt.title("Number of Countries by Region", size = 20)
plt.xticks(size = 15, rotation = "90")
plt.yticks(size = 15)
plt.xlabel(None)
plt.ylabel(None)
plt.show()
```

### Students record

```python
sr = pd.read_csv("student-alcohol-consumption.csv")
sr = sr[sr.columns.drop("Unnamed: 0")]
sr[:4]
```

```python
plt.figure(figsize = (18, 7))
sns.scatterplot(x = "absences", y = "G3", hue = "location", hue_order = ["Rural", "Urban"], data = sr, s = 100)
plt.title("Absences and G3 by Location", size = 20)
plt.xlabel("Absesces", color = "red", size = 20)
plt.ylabel("G3", color = "green", size = 20)
plt.xticks(size = 15)
plt.yticks(size = 15)
plt.show()
```

```python
# Create a count plot of school with location subgroups

plt.figure(figsize = (18, 7))
sns.countplot(x = "school", hue = "location", data = sr, palette = {"Rural": "green", "Urban": "blue"})
plt.title("School Students of classes by Location", size = 20)
plt.xlabel("Schools", color = "red", size = 20)
plt.ylabel("Frequency", color = "red", size = 20)
plt.xticks(size = 15)
plt.yticks(size = 15)
plt.show()
```

### Creating subplots with col and row

```python
sns.relplot(x = "absences", y = "study_time", data = sr, kind = "scatter", col = "study_time", color = "blue")
plt.show()
```

### Creating two-factor subplots

```python
plt.figure(figsize = (18, 7))
g = sns.relplot(x = "G1", y = "G3", data = sr, color = "deeppink", s = 100)
g.fig.set_size_inches(18, 7)
plt.title("First Term and Third Term Scores of Students", size = 20)
plt.xlabel("First Term", color = "red", size = 20)
plt.ylabel("Third Term", color = "red", size = 20)
plt.xticks(size = 15)
plt.yticks(size = 15)
plt.show()
```

**Adjust further to add subplots based on family support**

```python
g = sns.relplot(x = "G1", y = "G3", data = sr, row = "schoolsup", col_order = ["yes", "no"], kind = "scatter", s = 100, color = "black")
g.fig.set_size_inches(18, 7)
plt.show()
```

**Adjust further to add subplots based on family support (Columns)**

```python
g = sns.relplot(x = "G1", y = "G3", data = sr, col = "schoolsup", col_order = ["yes", "no"], 
                row = "famsup", row_order = ["yes", "no"], kind = "scatter", s = 100, color = "orange")
g.fig.set_size_inches(18, 7)
plt.show()
```

### Changing the size of scatter plot points

```python
mpg = pd.read_csv("mpg.csv")
mpg[:5]
```

**Horsepower and Mpg by Cylinders**

```python
g = sns.relplot(x = "horsepower", y = "mpg", s = 100, hue = "cylinders", data = mpg, palette = {3: "red", 4: "orange", 5: "blue", 6: "green", 8: "black"})
g.fig.set_size_inches(18, 7)
plt.title("Horsepower and Mpg by Cylinders", size = 25, color = "brown")
plt.xlabel("Horsepower", color = "brown", size = 25)
plt.ylabel("Mpg", color = "brown", size = 25)
plt.xticks(size = 15)
plt.yticks(size = 15)
print("\n")
plt.show()
```

**Acceleration and Mpg by Origin**

```python
g = sns.relplot(x = "acceleration", y = "mpg", hue = "origin", s = 100, kind = "scatter", data = mpg)
g.fig.set_size_inches(18, 7)
plt.title("Acceleration and Mpg by Origin", size = 25, color = "red")
plt.xlabel("Acceleration", color = "red", size = 25)
plt.ylabel("Mpg", color = "red", size = 25)
plt.xticks(size = 15)
plt.yticks(size = 15)
print("\n")
plt.show()
```

### Interpreting line plots

```python
g = sns.relplot(x = "model_year", y = "mpg", data = mpg, kind = "line")
g.fig.set_size_inches(18, 7)
plt.title("Miles Per Gallon by Model Year", size = 25, color = "blue")
plt.xlabel("Model Year", color = "blue", size = 25)
plt.ylabel("Mpg", color = "blue", size = 25)
plt.xticks(size = 15)
plt.yticks(size = 15)
plt.show()
```

**Plot with Standard Deviation**

```python
g = sns.relplot(x = "model_year", y = "mpg", data = mpg, kind = "line", ci = "sd", color = "lime")
g.fig.set_size_inches(18, 7)
plt.title("Miles Per Gallon by Model Year", size = 25, color = "green")
plt.xlabel("Model Year", color = "green", size = 25)
plt.ylabel("Mpg", color = "green", size = 25)
plt.xticks(size = 15)
plt.yticks(size = 15)
plt.show()
```

**Create line plot of model year vs. horsepower**

```python
g = sns.relplot(x = "model_year", y = "horsepower", kind = "line", ci = None,
                color = "orange", hue = "origin", linewidth = 3, data = mpg )
g.fig.set_size_inches(18, 7)
plt.title("Horsepower by Model Year", size = 25, color = "brown")
plt.xlabel("Model Year", color = "brown", size = 25)
plt.ylabel("Horse Power", color = "brown", size = 25)
plt.xticks(size = 15)
plt.yticks(size = 15)
plt.show()
```

### Count Plots

```python
g = sns.catplot(x = "Internet usage", data = servey, kind = "count", palette = "hsv")
g.fig.set_size_inches(18, 7)
plt.title("Internet usage by Students", size = 25, color = "red")
plt.xlabel("Internet Usage", color = "red", size = 25)
plt.ylabel("Studfents", color = "red", size = 25)
plt.xticks(size = 15)
plt.yticks(size = 15)
plt.show()
```

**Create a bar plot of interest in math, separated by gender**

```python
g = sns.catplot(x = "Gender", y = "Mathematics", data = servey, kind = "bar", palette = "rainbow")
g.fig.set_size_inches(18, 7)
plt.title("Interest in Mathematics by Gender", size = 25, color = "orange")
plt.xlabel("Gender", color = "orange", size = 25)
plt.ylabel("Mathematics", color = "orange", size = 25)
plt.xticks(size = 15)
plt.yticks(size = 15)
plt.show()
```

**Create bar plot of average final grade in each study category**

```python
g = sns.catplot(x = "study_time", y = "G3", data = sr, palette = "prism", ci = None,
                kind = "bar", order = ["<2 hours", "2 to 5 hours", "5 to 10 hours", ">10 hours"])
g.fig.set_size_inches(18, 7)
plt.title("Study category by average final grade", size = 25, color = "blue")
plt.xlabel("Study Category", color = "blue", size = 25)
plt.ylabel("Average Final Grade", color = "blue", size = 25)
plt.xticks(size = 15)
plt.yticks(size = 15)
plt.show()
```

### Create and interpret a box plot

```python
g = sns.catplot(x = "study_time", y = "G3", data = sr, palette = "summer_r",
                  kind = "box", order = ["<2 hours", "2 to 5 hours", "5 to 10 hours", ">10 hours"])
g.fig.set_size_inches(18, 7)
plt.title("Study category by average final grade", size = 25, color = "red")
plt.xlabel("Study Category", color = "red", size = 25)
plt.ylabel("Average Final Grade", color = "red", size = 25)
plt.xticks(size = 15)
plt.yticks(size = 15)
plt.show()
```

**final score through location by internet access**

```python
g = sns.catplot(x = "internet", y = "G3", hue = "location", data = sr,
                palette = "coolwarm", kind = "box")

g.fig.set_size_inches(18, 7)
plt.title("Final Score by Internet Access through Location", size = 25, color = "purple")
plt.xlabel("Internet Access", color = "purple", size = 25)
plt.ylabel("Average Final Grade", color = "purple", size = 25)
plt.xticks(size = 15)
plt.yticks(size = 15)
plt.show()
```

**Remove Outliers**

```python
g = sns.catplot(x = "internet", y = "G3", hue = "location", data = sr,
                palette = "coolwarm", kind = "box", sym = "")

g.fig.set_size_inches(18, 7)
plt.title("Final Score by Internet Access through Location", size = 25, color = "purple")
plt.xlabel("Internet Access", color = "purple", size = 25)
plt.ylabel("Average Final Grade", color = "purple", size = 25)
plt.xticks(size = 15)
plt.yticks(size = 15)
plt.show()
```

### Adjusting the whiskers

```python
g = sns.catplot(x = "romantic", y = "G3", data = sr, kind = "box", palette = "prism", whis = [5, 95])
g.fig.set_size_inches(18, 7)
plt.title("Final Score by Relationship", size = 25, color = "blue")
plt.xlabel("Relationship", color = "blue", size = 25)
plt.ylabel("Average Final Grade", color = "blue", size = 25)
plt.xticks(size = 15)
plt.yticks(size = 15)
plt.show()
```

### Customizing point plots

```python
g = sns.catplot(x = "famrel", y = "absences", data = sr, kind = "point", color = "green", linewidth = 3)
g.fig.set_size_inches(18, 7)
plt.title("Absences by Family Relationship", size = 25, color = "lime")
plt.xlabel("Family Relationship", color = "lime", size = 25)
plt.ylabel("Absences", color = "lime", size = 25)
plt.xticks(size = 15)
plt.yticks(size = 15)
plt.show()
```

### Point plots with subgroups

```python
g = sns.catplot(x = "romantic", y = "absences", data = sr, estimator = median,
                hue = "school", kind = "point", palette = "hot", ci = None)
g.fig.set_size_inches(18, 7)
plt.title("Absences by Romantic Relationship through Schools", size = 25, color = "blue")
plt.xlabel("Romantic Relationship", color = "blue", size = 25)
plt.ylabel("Absences", color = "blue", size = 25)
plt.xticks(size = 15)
plt.yticks(size = 15)
plt.show()
```
