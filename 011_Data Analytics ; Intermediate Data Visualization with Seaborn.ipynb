### Import Libraries

```python
import warnings
warnings.filterwarnings("ignore")
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Import Dataset

```python
sg = pd.read_csv("schoolimprovement2010grants.csv")
sg[:3]
```

```python
# Display a Seaborn distplot
plt.figure(figsize = (18, 7))
sns.distplot(sg['Award_Amount'], color = "blue", bins = 20)
plt.title("Distribution Plot of Award Amount", color = "red", size = 20)
plt.xticks(size = 15)
plt.xticks(size = 15)
plt.xlabel("Award Amount", size = 20, color = "red")
plt.ylabel("Density", size = 20, color = "red")
plt.show()
```

### Rug plot and kde shading

```python
plt.figure(figsize = (18, 7))
sns.distplot(sg['Award_Amount'],
             hist=False,
             rug=True,
             color = "red",
             kde_kws={'shade': True})
plt.title("Distribution Plot of Award Amount", color = "blue", size = 20)
plt.xticks(size = 15)
plt.xticks(size = 15)
plt.xlabel("Award Amount", size = 20, color = "blue")
plt.ylabel("Density", size = 20, color = "blue")
plt.show()
```

### Create a regression plot

```python
ip = pd.read_csv("insurance_premiums.csv")
ip[:5]
```

```python
plt.figure(figsize = (18, 7))
sns.regplot(x = "insurance_losses", y = "premiums", data = ip, color = "orange")
plt.title("Regression Plots of Premiums by Insurance Losses", color = "brown", size = 20)
plt.xticks(size = 15)
plt.xticks(size = 15)
plt.xlabel("Insurance Losses", size = 20, color = "brown")
plt.ylabel("Premiums", size = 20, color = "brown")
plt.show()
```

```python
plt.figure(figsize = (18, 7))
sns.lmplot(x = "insurance_losses", y = "premiums", data = ip, height=5, aspect=3)
plt.title("Regression Plots of Premiums by Insurance Losses", color = "blue", size = 20)
plt.xticks(size = 15)
plt.xticks(size = 15)
plt.xlabel("Insurance Losses", size = 20, color = "blue")
plt.ylabel("Premiums", size = 20, color = "blue")
plt.show()
```

```python
# Create a regression plot using hue
sns.lmplot(data=ip,
           x="insurance_losses",
           y="premiums",
           row="Region")

# Show the results
plt.show()
```

### Creating heatmaps

```python
ds = pd.read_csv("daily_show_guests_cleaned.csv")
ds[:3]
```

```python
# Create a crosstab table of the data
plt.figure(figsize = (18, 7))
pd_crosstab = pd.crosstab(ds["Group"], ds["YEAR"])

# Plot a heatmap of the table
sns.heatmap(pd_crosstab, cbar=False, cmap="BuGn", linewidths=0.3)

# Rotate tick marks for visibility
plt.yticks(rotation=0)
plt.xticks(rotation=90)

plt.show()
```
