### Field work and Desk Work Hours

```python
fd = pd.read_csv("field_desk.csv")
fd = 
```

```python
fd = fd.melt("officer", var_name = "Work", value_name = "Hours")
fd
```

```python
plt.figure(figsize = (18, 7))
graph = sns.barplot(x = "officer", y = "Hours", hue = "Work", data = fd, palette = "spring_r")
for p in graph.patches:
        graph.annotate('{:.01f}'.format(p.get_height()), 
                      (p.get_x()+0.2, p.get_height()),
                       ha='center', va='bottom',color= 'black', size = 15)
plt.title("Field work and Desk Work of Officers by Hours", size = 15)
plt.xticks(size = 15)
plt.yticks(size = 15)
plt.xlabel(None)
plt.ylabel(None)
plt.show()
```
